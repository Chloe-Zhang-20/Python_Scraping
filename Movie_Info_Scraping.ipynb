```python
import pandas as pd
columns=['Movie_name','Rating','Genre','Release_time','Review_media']
df=pd.DataFrame(columns=columns)

df
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
namelist=pd.read_csv('/Users/chloe/Desktop/namelist.csv')
namelist=namelist.values.tolist()
movies=[]
for name in namelist:
    movies.append(name[0])
movies
```

```python
headers = {
    "User-Agent": "Mozilla/4.0 (compatible; MSIE 5.5; Windows NT)",
           "Referer": "http://www.zhihu.com/articles"
}
```

```python
base_url='https://www.rottentomatoes.com/m/'

title=[]
rate=[]
release_time=[]





  
for m in movies:
    try: 
        r=requests.get(base_url+m,headers=headers)
        soup=BeautifulSoup(r.content, 'lxml')
        title=soup.find(class_='mop-ratings-wrap__title mop-ratings-wrap__title--top').contents[0]
        
    
        ratediv=soup.find(class_='mop-ratings-wrap__half audience-score')
        rate=ratediv.find(class_='mop-ratings-wrap__text--small').contents[0]
        rate=rate.split(': ')
        rate=rate[1]
    
    
        genre_list=soup.find(class_='meta-value genre').contents[0].split(',')
        genre=[]
        for g in genre_list:
            genre.append(g.strip('\n '))
  
    
        div=soup.find(class_='content-meta info')
        release_time=div.time.get_text().strip('\n ')
    
    
        review=soup.find_all(class_='unstyled articleLink fgm')
        review_media=[]
        for r in review:
            review_media.append(r.contents[0])
    
        
    
        df_length = len(df)
        df.loc[df_length] = [title,rate,genre,release_time,review_media] 


    
    except Exception as e:
        pass

    continue
    
print(df)
```

```python
file_path = r'./output.xlsx'
writer = pd.ExcelWriter(file_path)
df.to_excel(writer,index=False,encoding='utf-8',sheet_name='Sheet')
writer.save()
```
