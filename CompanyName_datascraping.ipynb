```python
f = open("company_list.txt")
import pandas as pd
company_list=pd.DataFrame(columns=['company_name','city','state','website'])
print(company_list)
for line in f:
    count=3
    i=len(line)-1
    while count>0:
        if line[i]==',':
            count-=1
        i-=1
    company_name=line[:i+1]
    city=line[i+2:].split(',')[0]
    state=line[i+2:].split(',')[1]
    website=line[i+2:].split(',')[2]
    df2 = pd.DataFrame([[company_name,city,state,website]], columns=['company_name','city','state','website'])
    #print(df2)
    company_list=pd.concat([company_list,df2],ignore_index = True)
```
